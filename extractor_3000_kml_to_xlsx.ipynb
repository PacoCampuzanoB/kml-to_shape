```python
import geopandas as gpd
import os
import pandas as pd
import fiona
from shapely.geometry import Point
import subprocess
from pykml import parser
from lxml import etree
```

```python
print(fiona.supported_drivers)
```

```python
ns = {'kml': 'http://www.opengis.net/kml/2.2'}
```

```python
with open("doc.kml", 'r', encoding='utf-8') as f:
    root = parser.parse(f).getroot()
```

```python
nombres = []
coordenadas = []
```

```python
for placemark in root.Document.Placemark:
    # Obtener el nombre del Placemark
    nombres.append(placemark.name.text)
```

```python
nombres = []
longitudes = []
latitudes = []
altitudes = []

claves = []
nombres_estac = []
estados = []
municipios = []
org_cuenca = []
cuencas = []
tipos_est = []
inicios = []
mas_recientes = []
situaciones = []
diarios = []
mensuales = []
extremos = []
normales_61_90 = []
normales_71_00 = []
normales_81_10 = []
normales_91_20 = []
```

```python
for placemark in root.findall('.//kml:Placemark', ns):
    # Obtener el nombre del Placemark
    name = placemark.find('.//kml:name', ns)
    if name is not None:
        nombres.append(name.text)

    # Obtener las coordenadas del Point
    coordinates = placemark.find('.//kml:Point/kml:coordinates', ns)
    if coordinates is not None:
        lon, lat, alt = map(float, coordinates.text.strip().split(','))
        longitudes.append(lon)
        latitudes.append(lat)
        altitudes.append(alt)
    
    # Extraer informaciÃ³n del ExtendedData
    clave = placemark.find(".//kml:SimpleData[@name='CLAVE']", ns)
    nombre_estac = placemark.find(".//kml:SimpleData[@name='NOMBRE']", ns)
    estado = placemark.find(".//kml:SimpleData[@name='ESTADO']", ns)
    municipio = placemark.find(".//kml:SimpleData[@name='MUNICIPIO']", ns)
    org_cuen = placemark.find(".//kml:SimpleData[@name='ORG_CUENCA']", ns)
    cuenca = placemark.find(".//kml:SimpleData[@name='CUENCA']", ns)
    tipo_est = placemark.find(".//kml:SimpleData[@name='TIPO_EST']", ns)
    inicio = placemark.find(".//kml:SimpleData[@name='INICIO']", ns)
    mas_reciente = placemark.find(".//kml:SimpleData[@name='MAS_RECIENTE']", ns)
    situacion = placemark.find(".//kml:SimpleData[@name='SITUACION']", ns)
    diario = placemark.find(".//kml:SimpleData[@name='DIARIOS']", ns)
    mensual = placemark.find(".//kml:SimpleData[@name='MENSUALES']", ns)
    extremo = placemark.find(".//kml:SimpleData[@name='EXTREMOS']", ns)
    normal_61_90 = placemark.find(".//kml:SimpleData[@name='NORMALES_1961_1990']", ns)
    normal_71_00 = placemark.find(".//kml:SimpleData[@name='NORMALES_1971_2000']", ns)
    normal_81_10 = placemark.find(".//kml:SimpleData[@name='NORMALES_1981_2010']", ns)
    normal_91_20 = placemark.find(".//kml:SimpleData[@name='NORMALES_1991_2020']", ns)
    
    claves.append(clave.text if clave is not None else None)
    nombres_estac.append(nombre_estac.text if nombre_estac is not None else None)
    estados.append(estado.text if estado is not None else None)
    municipios.append(municipio.text if municipio is not None else None)
    org_cuenca.append(org_cuen.text if org_cuen is not None else None)
    cuencas.append(cuenca.text if cuenca is not None else None)
    tipos_est.append(tipo_est.text if tipo_est is not None else None)
    inicios.append(inicio.text if inicio is not None else None)
    mas_recientes.append(mas_reciente.text if mas_reciente is not None else None)
    situaciones.append(situacion.text if situacion is not None else None)
    diarios.append(diario.text if diario is not None else None)
    mensuales.append(mensual.text if mensual is not None else None)
    extremos.append(extremo.text if extremo is not None else None)
    normales_61_90.append(normal_61_90.text if normal_61_90 is not None else None)
    normales_71_00.append(normal_71_00.text if normal_71_00 is not None else None)
    normales_81_10.append(normal_81_10.text if normal_81_10 is not None else None)
    normales_91_20.append(normal_91_20.text if normal_91_20 is not None else None)
```

```python
print("claves -->", len(claves))
print("nombres_estac -->", len(nombres_estac))
print("estados -->", len(estados))
print("municipios -->", len(municipios))
print("org_cuenca -->", len(org_cuenca))
print("cuencas -->", len(cuencas))
print("tipos_est -->", len(tipos_est))
print("inicios -->", len(inicios))
print("mas_recientes -->", len(mas_recientes))
print("situaciones -->", len(situaciones))
print("diarios -->", len(diarios))
print("mensuales -->", len(mensuales))
print("extremos -->", len(extremos))
print("normales_61_90 -->", len(normales_61_90))
print("normales_71_00 -->", len(normales_71_00))
print("normales_81_10 -->", len(normales_81_10))
print("normales_91_20 -->", len(normales_91_20))
```

```python
df = pd.DataFrame({
    'clave': claves,
    'nombre': nombres_estac,
    'entidad_monitoreo': estados,
    'municipio_monitoreo': municipios,
    'org_cuenca': org_cuenca,
    'cuenca': cuencas,
    'tipo_estacion': tipos_est,
    'inicio': inicios,
    'mas_reciente': mas_recientes,
    'situacion': situaciones,
    'val_diarios': diarios,
    'val_mens': mensuales,
    'val_extr': extremos,
    'norm_61_90': normales_61_90,
    'norm_71_00': normales_71_00,
    'norm_81_10': normales_81_10,
    'norm_91_10': normales_91_20,
    'x_long': longitudes,
    'y_lat': latitudes,
    'altitud': altitudes
})
```
